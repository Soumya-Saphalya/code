```python

def convert(obj) :
    L  = []
    for i in json.loads(obj):
    
        L.append(i['name'])
    return L     
```

used json.loads for converting string of lists to list

```python
movies['genres'] =  movies['genres'].apply(convert)
```

```python
movies.keywords = movies.keywords.apply(convert)
```

```python
def convert2(obj) :
    L = []
    count = 0 
    for i in json.loads(obj):
         
        if(count<5) :
            L.append(i['name'])
            count += 1
        else :
            break
    return L     
```

```python
movies.cast = movies.cast.apply(convert2)
```

```python
def convert3(obj) :
    l = []
    
    for i in json.loads(obj) :
        
        if(i['job'] ==  "Director") :
            
            l.append(i['name'])
            
            break
            
        
        
    return l       
```

```python
movies.crew = movies.crew.apply(convert3)
```

```python
movies.genres = movies.genres.apply(lambda x :[i.replace(" ","")for i in x])
movies.keywords = movies.keywords.apply(lambda x :[i.replace(" ","")for i in x])
movies.crew = movies.crew.apply(lambda x :[i.replace(" ","")for i in x])
movies.cast = movies.cast.apply(lambda x :[i.replace(" ","")for i in x])
movies.overview = movies.overview.apply(lambda x :[i.replace(" ","")for i in x])
```

```python
new_movies.tags = new_movies.tags.apply(lambda x :" ".join(x))
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text) :
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_movies.tags.apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv  = CountVectorizer(stop_words= 'english',max_features = 5000)
```

```python
vectors = cv.fit_transform(new_movies['tags'])
vector = np.array(vectors)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse = True ,key = lambda x :x[1])[1:6]
```

```python
movie_index = new_movies[new_movies['title'] == 'Avatar'].index[0]
distances = sorted(list(enumerate(similarity[movie_index])),reverse = True ,key = lambda x :x[1])[1:6]

for i in distances :
        print(new_movies.iloc[i[0]].title)
```

we create a list of tupples with index as movie id and distances of cosine angles of vectors 

```python
similarity[1]
```

```python
def recommend(movie) :
    
    movie_index = new_movies[new_movies['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[movie_index])),reverse = True ,key = lambda x :x[1])[1:6]

    for i in distances :
        print(new_movies.iloc[i[0]].title)
    return    
    
```

```python
recommend('Avatar')
```
